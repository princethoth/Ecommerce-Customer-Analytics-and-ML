```python
# CLEAN SALES DATA
```

```python
# Drop the completely empty column
sd = sd.drop('Unnamed: 4', axis=1)
```

```python
# Remove rows with missing values
sd_clean = sd.dropna()
```

```python
# Filter to purchases only (exclude views and likes)
sd_clean = sd_clean[sd_clean['Interaction type'] == 'purchase']
```

```python
# Convert timestamp to datetime
sd_clean['Time stamp'] = pd.to_datetime(sd_clean['Time stamp'], format='%d/%m/%Y %H:%M')
```

```python
# Rename columns
sd_clean.columns = ['user_id', 'product_id', 'interaction_type', 'timestamp']
```

```python
# Convert user_id to int
sd_clean['user_id'] = sd_clean['user_id'].astype(int)
```

```python
print(f"Final shape: {sd_clean.shape}")
```

```python
#CLEAN CUSTOMER DETAILS
```

```python
# Rename columns (make consistent)
cd.columns = [
    'customer_id', 'age', 'gender', 'item_purchased', 'category',
    'purchase_amount', 'location', 'size', 'color', 'season',
    'review_rating', 'subscription_status', 'shipping_type',
    'discount_applied', 'promo_code_used', 'previous_purchases',
    'payment_method', 'frequency_of_purchases'
]
```

```python
#CLEAN PRODUCT DETAILS
```

```python
# Keep only useful columns
useful_columns = [
    'Uniqe Id', 'Product Name', 'Category', 'Selling Price',
    'About Product', 'Product Specification', 'Is Amazon Seller'
]

products_clean = products[useful_columns].copy()
```

```python
# Rename columns
products_clean.columns = [
    'product_id', 'product_name', 'category', 'selling_price',
    'about_product', 'product_specification', 'is_amazon_seller'
]
```

```python
# Clean selling price (remove $ and convert to float)
products_clean['selling_price'] = products_clean['selling_price'].str.replace('$', '').str.replace(',', '')
products_clean['selling_price'] = pd.to_numeric(products_clean['selling_price'], errors='coerce')
```

```python
# Handle missing categories
products_clean['category'] = products_clean['category'].fillna('Unknown')
```

```python
# Extract main category
products_clean['main_category'] = products_clean['category'].str.split('|').str[0].str.strip()
```

```python
# Drop rows with missing product_id or product_name
products_clean = products_clean.dropna(subset=['product_id', 'product_name'])
```

```python
print(f"CLEAN DATASETS:")
print(f"Sales (Purchases only): {sd_clean.shape}")
print(f"Customers: {cd.shape}")
print(f"Products: {products_clean.shape}")
```

```python
# Check date range
print(f"TIME PERIOD:")
print(f"Start: {sd_clean['timestamp'].min()}")
print(f"End: {sd_clean['timestamp'].max()}")
print(f"Duration: {(sd_clean['timestamp'].max() - sd_clean['timestamp'].min()).days} days")
```

```python
# Save cleaned data
sd_clean.to_csv('sales_data_clean.csv', index=False)
cd.to_csv('customers_clean.csv', index=False)
products_clean.to_csv('products_clean.csv', index=False)
```
